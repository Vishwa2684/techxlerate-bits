```python

from transformers import pipeline
import warnings

warnings.filterwarnings('ignore')

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli",device=-1)


text = "I hate you"
labels = ["fear", "loneliness", "joy", "anger", "depression", "anxiety", "love","sadness"]


results = classifier(text, candidate_labels=labels)

l=[]
v=[]
j=0
for i in results['labels']:
    l.append(i)
    v.append(results['scores'][j])
    j+=1
my_dict = dict(zip(l,v))
print(my_dict)
plt.figure(figsize=(8,8))
plt.pie(v,labels=l)
plt.title("Emotion chart")
plt.legend()
plt.show()

```
